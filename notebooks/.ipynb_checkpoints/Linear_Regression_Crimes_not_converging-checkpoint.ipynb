```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import pymc3 as pm
import arviz as az
```

```python
df = pd.read_csv('../data/us_state_crime.csv')
df.rename(columns={'crimes_per_1000':'crime'}, inplace=True)

useful_cols = ['state', 'crime', 'pct_poverty_rate', 'pct_single_parents']

df.head()
```

```python
plt.plot(df['pct_poverty_rate'], df['crime'], '.')
plt.xlabel('Poverty Rate for US States')
plt.ylabel('Crimes per 1000 ppl')
```

As the plot above shows, it is not easy to "draw a line of best fit.' The crime rate is scattered for the 50 states. 
We could try and fit a linear regression model. But perhaps, a Bayesian model is even more relevant here. We want to understand the *range* of certain parameters.

```python
df[useful_cols].describe()
```

```python
MAX_BASE = 50 #30 might be enough
MAX_POV_RATE = 30
MAX_PARENTS = 25
MAX_SD = 8
```

```python
crime = df['crime'].values
single_parents = df['pct_single_parents'].values
poverty = df['pct_poverty_rate'].values


with pm.Model() as crime_model:
    
    #define priors
    base_crime = pm.Uniform('base_crime', 0, MAX_BASE) # This is the price for a home with ZERO bedrooms
    parents_coeff = pm.Uniform('parents_coeff', 0, MAX_PARENTS) #Additional cost for each bedroom 
    pov_coeff = pm.Uniform('pov_coeff', 0, MAX_POV_RATE) #Additional cost for each bedroom 

    sigma = pm.HalfNormal('sigma', sd=MAX_SD)

    y_predicted = base_crime + parents_coeff * single_parents + pov_coeff * poverty 
    
    #define Likelihood
    likelihood = pm.Normal('y',
                           mu = y_predicted, sigma = sigma,
                           observed = crime)
        
```

```python
crime_model
```

```python
len(crime) # Number of observations we have, 50 US States + DC
```

# Ready to Start Sampling

We are going to sample (draw from) 10,000 times. Each attempt will get us one set of (coefficients & sigma).
- If the error is "acceptable," we will randomly sample more in that neighborhood of m, x and sigam.
- If the error is large, we will 'jump' elsewhere and try sampling there.
- Note: the first 5000 values will be discarded. They are to warm-up the model.

```python
with crime_model:
    trace = pm.sample(10000, 
                      chains=2,
                      tune=5000)
```

### What do the results look like?

```python
pm.summary(trace).round(2)
```

### Let us look at the Posterior

```python
pm.traceplot(trace);
axes = pm.plot_posterior(trace)
```

```python
pm.forestplot(trace);
```

```python
az.plot_pair(trace, var_names=['parents_coeff', 'pov_coeff'], plot_kwargs={'alpha': 0.1})
```

```python
plt.plot(poverty, crime, 'C0.')

c_pov = trace['pov_coeff'].mean()
intercept = trace['base_price'].mean()

draws = range(0, len(trace['']), 10)
plt.plot(poverty, trace['α'][draws] + trace['β'][draws]
         * poverty[:, np.newaxis], c='gray', alpha=0.5)

plt.plot(poverty, intercept + c_pov * poverty, c='k',
         label=f'crime = {intercept:.2f} + {c_pov:.2f} * poverty')

plt.xlabel('Poverty Rate')
plt.ylabel('crime', rotation=0)
plt.legend()
```
